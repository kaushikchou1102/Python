```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example dataset
tips = sns.load_dataset('tips')

# Scatter plot using Seaborn
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

```python
import warnings
# Suppressing warnings
warnings.filterwarnings('ignore')

# Example dataset
fmri = sns.load_dataset('fmri')

# Line plot using Seaborn
sns.lineplot(x='timepoint', y='signal', data=fmri)
plt.title('Line Plot of Signal over Time')
plt.show()
```

```python
# Histogram using Seaborn
sns.histplot(data=tips, x='total_bill', bins=20, kde=True)
plt.title('Histogram of Total Bill')
plt.show()
```

```python
# Bar plot using Seaborn
sns.barplot(x='day', y='total_bill', data=tips, estimator=sum)
plt.title('Bar Plot of Total Bill by Day')
plt.show()
```

```python
# Box plot using Seaborn
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot of Total Bill by Day')
plt.show()
```

```python
# Violin plot using Seaborn
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Violin Plot of Total Bill by Day')
plt.show()
```

```python
# Heatmap using Seaborn
flights = sns.load_dataset('flights')
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(flights_pivot, cmap='YlGnBu', linecolor='white', linewidth=1)
plt.title('Heatmap of Flight Passengers')
plt.show()
```

```python
# Pair plot using Seaborn
sns.pairplot(tips, hue='sex')
plt.title('Pair Plot of Tips Dataset')
plt.show()
```

```python
# Joint plot using Seaborn
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
plt.suptitle('Joint Plot of Total Bill vs Tip')
plt.show()
```

```python
# Facet grid using Seaborn
g = sns.FacetGrid(tips, col='time', row='smoker')
g.map(sns.scatterplot, 'total_bill', 'tip')

plt.suptitle('Facet Grid of Total Bill vs   Tip')
plt.show()
```

```python
# Count plot using Seaborn
sns.countplot(x='day', data=tips)
plt.title('Count Plot of Days')
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Demo data
theta = np.linspace(0, 2*np.pi, 100)
r = np.sin(theta)

# Polar plot
fig = go.Figure(go.Scatterpolar(
    r=r,
    theta=theta,
    mode='lines',
    title='Polar Plot Example'
))
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Demo data
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2)

# Contour plot
fig = go.Figure(go.Contour(
    z=Z,
    x=x,
    y=y,
    colorscale='Viridis',
    title='Contour Plot Example'
))
fig.show()
```

```python
import plotly.graph_objects as go

# Demo data
nodes = [{'id': 'A'}, {'id': 'B'}, {'id': 'C'}]
edges = [{'source': 'A', 'target': 'B'}, {'source': 'B', 'target': 'C'}]

# Network graph
fig = go.Figure(go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=[n['id'] for n in nodes],
    ),
    link=dict(
        source=[e['source'] for e in edges],
        target=[e['target'] for e in edges],
    ),
    title='Network Graph Example'
))
fig.show()
```

```python
import plotly.graph_objects as go

# Demo data
events = ['Event A', 'Event B', 'Event C']
dates = ['2023-01-01', '2023-03-01', '2023-06-01']

# Timeline plot
fig = go.Figure()
for i, event in enumerate(events):
    fig.add_trace(go.Scatter(x=[dates[i], dates[i]], y=[0, 1], mode='lines', name=event))
fig.update_layout(title='Timeline Plot Example', xaxis_title='Date')
fig.show()
```
